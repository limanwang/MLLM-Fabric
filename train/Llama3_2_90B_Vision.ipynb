### Data Prep
We'll be using a sampled version of the ROCO radiography dataset. You can access the dataset [here](https://huggingface.co/datasets/unsloth/Radiology_mini). The full dataset is [here](https://huggingface.co/datasets/eltorio/ROCOv2-radiology).

The dataset includes X-rays, CT scans and ultrasounds showcasing medical conditions and diseases. Each image has a caption written by experts describing it. The goal is to finetune a VLM to make it a useful analysis tool for medical professionals.


````python
import os
from PIL import Image

def concat_images(image_paths):
    """
    image_paths: A list of 10 image paths in order.
    """
    assert len(image_paths) == 10, "Expected 10 images as input"

    # Process the first 5 images for the top row
    upper_imgs = [Image.open(p.replace("\\", "/")) for p in image_paths[:5]]
    lower_imgs = [Image.open(p.replace("\\", "/")) for p in image_paths[5:]]

    # Resize all to the same height (smallest one), then concatenate horizontally
    min_height_upper = min(img.height for img in upper_imgs)
    upper_imgs_resized = [img.resize(
        (int(img.width * min_height_upper / img.height), min_height_upper)
    ) for img in upper_imgs]
    upper_concat = concat_horizontal(upper_imgs_resized)

    min_height_lower = min(img.height for img in lower_imgs)
    lower_imgs_resized = [img.resize(
        (int(img.width * min_height_lower / img.height), min_height_lower)
    ) for img in lower_imgs]
    lower_concat = concat_horizontal(lower_imgs_resized)

    # Resize both rows to the same width (largest one), then concatenate vertically
    max_width = max(upper_concat.width, lower_concat.width)
    upper_concat = upper_concat.resize((max_width, upper_concat.height))
    lower_concat = lower_concat.resize((max_width, lower_concat.height))
    final_image = concat_vertical([upper_concat, lower_concat])

    return final_image

def concat_horizontal(images):
    total_width = sum(img.width for img in images)
    max_height = max(img.height for img in images)
    new_img = Image.new('RGB', (total_width, max_height))
    x_offset = 0
    for img in images:
        new_img.paste(img, (x_offset, 0))
        x_offset += img.width
    return new_img

def concat_vertical(images):
    max_width = max(img.width for img in images)
    total_height = sum(img.height for img in images)
    new_img = Image.new('RGB', (max_width, total_height))
    y_offset = 0
    for img in images:
        new_img.paste(img, (0, y_offset))
        y_offset += img.height
    return new_img
````

````python
import os
from PIL import Image
import json

def convert_sharegpt_to_unsloth_with_cache(sample, image_cache_dir="temp_concat"):
    """
    Process ShareGPT formatted data. Load the original image (if exists) and copy it to a cache directory (optional).
    """
    os.makedirs(image_cache_dir, exist_ok=True)

    # The image is a single one, use it directly
    image_path = sample["image"][0]
    # print(image_path)
    assert isinstance(image_path, str), "sample['image'] should be a string path"

    # Construct the cached output path (optionally copy to image_cache_dir)
    image_cache_dir = "/workspace/train_img/output_images_train4000/"
    output_path = os.path.join(image_cache_dir, image_path)
    # print(output_path)
    if os.path.exists(output_path):
        try:
            _ = Image.open(output_path).convert("RGB")
            # print(f"[Info] Cached image found: {output_path}")
        except Exception as e:
            print(f"[Warning] Failed to load cached image: {output_path}, will re-copy.")
            # _ = save_image_copy(image_path, output_path)
    # else:
        # _ = save_image_copy(image_path, output_path)
    
    # Assemble conversation data
    user_content = []
    for conv in sample["conversations"]:
        if conv["from"] == "human":
            user_content.append({"type": "text", "text": conv["value"]})
            break
    # Only include the concatenated image
    user_content.append({"type": "image", "image": output_path})

    assistant_content = []
    for conv in sample["conversations"]:
        if conv["from"] == "gpt":
            assistant_content.append({"type": "text", "text": conv["value"]})
            break

    return {
        "messages": [
            {"role": "user", "content": user_content},
            {"role": "assistant", "content": assistant_content}
        ]
    }

def generate_and_save_concat(sample, output_path):
    """
    Concatenate images and save the result (internal function).
    """
    image_paths = sample["image"]
    concat_img = concat_images(image_paths)
    concat_img.save(output_path)
    print(f"[Info] Saved new concat image: {output_path}")
    return concat_img
````

Before we do any finetuning, maybe the vision model already knows how to analyse the images? Let's check if this is the case!

````python
FastVisionModel.for_inference(model) # Enable for inference!

image_path = converted_dataset[0]["messages"][0]["content"][1]["image"]
image = Image.open(image_path).convert("RGB")
instruction = "You are a Fabric expert. Describe accurately what you see in this image."

messages = [
    {"role": "user", "content": [
        {"type": "image"},
        {"type": "text", "text": instruction}
    ]}
]
input_text = tokenizer.apply_chat_template(messages, add_generation_prompt = True)
inputs = tokenizer(
    image,
    input_text,
    add_special_tokens = False,
    return_tensors = "pt",
).to("cuda")

from transformers import TextStreamer
text_streamer = TextStreamer(tokenizer, skip_prompt = True)
_ = model.generate(**inputs, streamer = text_streamer, max_new_tokens = 128,
                   use_cache = True, temperature = 1.5, min_p = 0.1)
````

<a name="Train"></a>
### Train the model
Now let's use Huggingface TRL's `SFTTrainer`! More docs here: [TRL SFT docs](https://huggingface.co/docs/trl/sft_trainer). We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`. We also support TRL's `DPOTrainer`!

We use our new `UnslothVisionDataCollator` which will help in our vision finetuning setup.

````python
from unsloth import is_bf16_supported
from unsloth.trainer import UnslothVisionDataCollator
from trl import SFTTrainer, SFTConfig

FastVisionModel.for_training(model) # Enable for training!

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    data_collator = UnslothVisionDataCollator(model, tokenizer), # Must use!
    train_dataset = converted_dataset,
    args = SFTConfig(
        per_device_train_batch_size = 8,
        gradient_accumulation_steps = 2,
        warmup_steps = 20,
        # max_steps = 30,
        num_train_epochs = 1, # Set this instead of max_steps for full training runs
        learning_rate = 2e-4,
        fp16 = not is_bf16_supported(),
        bf16 = is_bf16_supported(),
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none",     # For Weights and Biases

        # You MUST put the below items for vision finetuning:
        remove_unused_columns = False,
        dataset_text_field = "",
        dataset_kwargs = {"skip_prepare_dataset": True},
        dataset_num_proc = 4,
        max_seq_length = 2048,
    ),
)
````

````python
# @title Show current memory stats
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")
````

````python
trainer_stats = trainer.train()
````

````python
# @title Show final memory and time stats
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory / max_memory * 100, 3)
lora_percentage = round(used_memory_for_lora / max_memory * 100, 3)
print(f"{trainer_stats.metrics['train_runtime']} seconds used for training.")
print(
    f"{round(trainer_stats.metrics['train_runtime']/60, 2)} minutes used for training."
)
print(f"Peak reserved memory = {used_memory} GB.")
print(f"Peak reserved memory for training = {used_memory_for_lora} GB.")
print(f"Peak reserved memory % of max memory = {used_percentage} %.")
print(f"Peak reserved memory for training % of max memory = {lora_percentage} %.")
````
